```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
norm_data = data2.copy()
def minmaxscaler(x):
    for columnName, columnData in x.iteritems():
        x[columnName] = scaler.fit_transform(np.array(columnData).reshape(-1, 1))
    
minmaxscaler(norm_data)
norm_data.head()
```

```python
km = list(range(2,11))
sum_of_squared_distances = []
for i in km:
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(norm_data)
    sum_of_squared_distances.append(kmeans.inertia_)

plt.figure(figsize=(10, 5))
plt.plot(km, sum_of_squared_distances, 'go--')
plt.xlabel('Number of Clusters')
plt.ylabel('Within Cluster Sum of squares')
plt.title('Elbow Curve to find optimum K')
```

```python
# Instantiating
kmeans4 = KMeans(n_clusters = 5)

# Training the model
kmeans4.fit(norm_data)

# predicting
y_pred = kmeans4.fit_predict(norm_data)
print(y_pred)

data2
```

```python
data2['2020'] = y_pred+1
```

```python
centroids = kmeans4.cluster_centers_
centroids = pd.DataFrame(centroids, columns=['Country_name','Country Code','1960','2019','2021'])
centroids.index = np.arange(1, len(centroids)+1)
centroids
```
